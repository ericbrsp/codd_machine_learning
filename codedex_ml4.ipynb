## Aprendizagem não supervisionada
Agora que exploramos o aprendizado supervisionado, vamos dar uma olhada no aprendizado não supervisionado.

Se você já assistiu à Netflix ou à HBO, esses serviços de streaming oferecem milhares de filmes e séries, mas só conseguem exibir alguns na sua página inicial. Como eles decidem quais mostrar? O sistema de recomendações agrupa os usuários com base nos hábitos de visualização – se você assiste a séries semelhantes às de outra pessoa, provavelmente também gostará dos favoritos dela.

Este é um exemplo clássico de aprendizado não supervisionado . Ao contrário do aprendizado supervisionado, que prevê uma meta específica (como "um usuário gostará deste programa?"), o aprendizado não supervisionado revela padrões e agrupamentos em dados sem rótulos. Cabe a nós, os programadores, decidir como usar esses insights.

Aqui estão alguns exemplos de aprendizado não supervisionado:

- Segmentação de clientes : agrupe os clientes por hábitos de compra para que as empresas possam adaptar estratégias de marketing para cada grupo.
- Detecção de Anomalias : Encontre comportamentos incomuns, como uma falha do sistema, usando agrupamento. Isso é aprendizado não supervisionado porque os dados não são explicitamente rotulados como normal/ failure; em vez disso, um modelo de agrupamento detecta quando um ponto de dados está fora dos padrões normais.
- Modelagem de Tópicos : Se você tiver milhares de artigos sem tags em tópicos específicos, um modelo de ML não supervisionado poderia agrupá-los com base em seu conteúdo. Uma vez formados os grupos, o engenheiro poderia investigá-los e atribuir a cada um deles o rótulo apropriado.


## Problema de agrupamento
Um dos algoritmos de aprendizado não supervisionados é o agrupamento.

Agrupamento consiste em agrupar pontos de dados semelhantes entre si em "clusters", sem conhecer previamente os rótulos corretos. O algoritmo procura padrões nos dados e os organiza de forma que pontos no mesmo cluster sejam mais semelhantes do que pontos em clusters diferentes.

Imagine que você queira encontrar cidades americanas com climas semelhantes para suas próximas férias. Suponha que você escolha dois recursos:

- average_humidity
- average_temperature

Você poderia plotar cidades em um gráfico 2D e provavelmente identificar aglomerados:

- Cidades quentes e úmidas: Miami, Houston
- Cidades quentes e secas: Phoenix, Las Vegas
- Cidades amenas: Seattle, SF, Boston, Chicago, Minneapolis, Denver

Mas e se adicionássemos um terceiro recurso, como average_rainfall, então seria:

- average_humidity
- average_temperature
- average_rainfall

Ainda é possível visualizar em um gráfico 3D, mas fica mais difícil ver os clusters claramente.

Agora, e se você adicionar um quarto ou quinto recurso? Humanos não conseguem visualizar facilmente além do 3D, mas algoritmos de agrupamento podem lidar com muitas dimensões (chamamos isso de espaço N-dimensional ). 🤯 Este conceito é fundamental em muitos modelos de ML que você verá ao longo deste curso.
